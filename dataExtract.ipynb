## Imports

```python
import pandas as pd
```

```python
import json
```

## Making a dataframe from a csv file

```python
df = pd.read_csv('./nasdaq_screener_1695101090068.csv')
```

```python
df
```

## Selecting only the symbol and name column, and converting it to a python dictionary

```python
dict = df.loc[:,["Symbol","Name"]].to_dict()
dict
```

## Creating an array, and populating it with the data from the dictionary

```python
array = []
```

```python
for symbol, name in zip(list(dict['Symbol'].values()), list(dict['Name'].values())):
    array.append({'symbol': str(symbol), 'name': str(name)})
    
```

```python
array
```

## Exporting the data as json

```python
data = json.dumps(array, indent=4)
```

```python
with open('data.json', 'w') as file:
    file.write(data)
```
